## 4.4 ТАБУЛИРОВАНИЕ ФУНКЦИЙ



```python
from math import sin, cos, radians
from pandas import DataFrame, set_option

a = float(input('Введите начальное значение интервала (в градусах): '))
b = float(input('Введите конечное значение интервала (в градусах): '))
h = float(input('Введите значение шага (в градусах): '))

def F(x_deg):
    """Вычисляем F(x) = 7 * sin^2(x) - 0.5 * cos(x), x в градусах"""
    x_rad = radians(x_deg)
    return 7 * sin(x_rad)**2 - 0.5 * cos(x_rad)

data = []
x = a
while x <= b + h/2:
    fx = F(x)
    data.append([f'{x:.2f}°', f'{fx:.6f}'])
    x += h

df = DataFrame(data, columns=['x (градусы)', 'F(x)'])
set_option('display.max_rows', None)
set_option('display.expand_frame_repr', False)

print('\n' + '-' * 60)
print('Значения функции F(x) = 7·sin²(x) - ½·cos(x) (x в градусах)')
print('-' * 60)
print(df.to_string(index=False))
```

## 4.5 ВВОД И ОБРАБОТКА ПОСЛЕДОВАТЕЛЬНОСТЕЙ


```python
digit = int(input("Введите число: "))

total = 0

# Вводим числа пока не надоест
print("Вводите числа (нажмите 0 что бы остановиться):")

while True:
    try:
        number = int(input("Число: "))
    except ValueError:
        print("Ошибка! Нажмите 0 что бы выйти")
        continue
    
    if number == 0:
        break
    
    if number % 10 == digit:
        total = total + number

print("Сумма:", total)
```
